```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
sr = cv.imread('/content/gdrive/MyDrive/roadline.jpeg')
plt.imshow(sr)
```

```python
#vertical line detection
kernel1 = np.array([[-1, 2, -1],
                    [-1, 2, -1],
                    [-1, 2, -1]])
vertical = cv.filter2D(src=sr, ddepth=-1, kernel=kernel1)
plt.imshow(vertical)
```

```python
#Horizontal Line Detection
kernel2 = np.array([[-1, -1, -1],
                    [2, 2, 2],
                    [-1, -1, -1]])
horizontal = cv.filter2D(src=sr, ddepth=-1, kernel=kernel2)
plt.imshow(horizontal)
```
